```python
# outlier removal

q1 = df.culmen_length_mm.quantile(0.25)
q3 = df.culmen_length_mm.quantile(0.75)
IQR = q3-q1
upper_limit = q3+1.5*IQR
lower_limit =q1-1.5*IQR
```

```python
df.median()
```

```python
df['culmen_length_mm'] = np.where(df['culmen_length_mm']>upper_limit,44,df['culmen_length_mm'])
```
